```python
###define a function that will extract float from string
import re
def extract_float(to_extract):
    if type(to_extract) == str:
        return re.findall(r"[-+]?\d*\.\d+|\d+", to_extract)[0]
    else:
        return to_extract
```

```python
### 在两个dataframe处理数据要小心，iat留言哦
def match_between_dataframe(marketMapping, productMapping, publicTime, all_prediction):
    return all_prediction.loc[(all_prediction['市场名称映射值'] == marketMapping) & (all_prediction['农产品名称映射值'] == productMapping) & 
                      (all_prediction['数据发布时间'] == publicTime)]['avgprice'].iat[0]
```

```python
test['avgprice']=test.apply(lambda row: match_between_dataframe(row['市场名称映射值'], row['农产品名称映射值'], row['数据发布时间'], all_prediction),axis=1)
```
